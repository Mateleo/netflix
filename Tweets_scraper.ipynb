## Libraries

```python
# Imports
import snscrape.modules.twitter as tweetscrape
import pandas as pd
from datetime import datetime
```

## Tweets Scraper by snscrape

```python
#Setting search's term
term1 = '#netflix bad movie until:2022-10-12 since:2012-10-12 lang:en'
term2 = '#netflix good movie until:2022-10-12 since:2012-10-12 lang:en'

# Setting maximum tweet's length
Max = 50000

# Creating list to append tweet data to
List_Tweet1 = []
List_Tweet2 = []
```

```python
# Define functions 
def tweetSraping(query, limit):
    """
    This function searchs in the twitter and returns 
    all tweets that have the desired query.
    """
    #Setting search's term
    term1 = query

    # Setting maximum tweet's length
    Max = limit
   
    # Creating list to append tweet data to
    List_Tweet1 = []

    #Importing the datetime to calculate the time for scraping the 50000 tweets
    start_time = datetime.now()
    
    # Using TwitterSearchScraper to scrape data and append the tweets in english to a list
    for i,tweet in enumerate(tweetscrape.TwitterSearchScraper(term1).get_items()):

        if i>Max:
            # Showing the attributes of one tweet to decide which are useful for later
            #print(vars(tweet))
            # break
            break
        List_Tweet1.append([tweet.date,tweet.id,tweet.content, tweet.lang, tweet.user.username, tweet.user.id, tweet.user.location])
    end_time = datetime.now()
    
    #Printing the time duration for scraping all the scraped tweets 
    print('Duration of scraping {} is: {}'.format(Max,end_time - start_time))
    
    return List_Tweet1

def tweetSave(tweets_list):
    """
    This function will save the created list .
    """
    # Creating a dataframe from the list of tweets generated by the function tweetSraping()
    data = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet_Id', 'Content', 'Language','Username','User_id','User_location'])
    
    return data
```

```python
# Extract tweets
# first one #netflix bad movie
List_Tweet1 = tweetSraping(term1,Max)
# second one #netflix good movie
List_Tweet2 = tweetSraping(term2,Max)
```

```python
# Creat dataframe 
data1 = tweetSave(List_Tweet1)
data2 = tweetSave(List_Tweet2)
```
